# Switch Call: Creation of Notebooks

```python

import os

source_tech = dbutils.widgets.get("source_tech")
catalog = dbutils.widgets.get("catalog")
schemas = dbutils.widgets.get("schemas")  # e.g. "schema1,schema2,schema3"
foundation_model = dbutils.widgets.get("foundation_model")
switch_config_path = dbutils.widgets.get("switch_config_path")
folder_name = dbutils.widgets.get("folder_name")

schema_list = [s.strip() for s in schemas.split(",") if s.strip()]

for schema in schema_list:
    input_dir = f"/Volumes/dbe_dbx_internships/{schema}/{schema}_volume/{folder_name}"
    output_folder = f"/Workspace/Users/depoplimontj@delawareconsulting.com/delaware_lakebridge/switch_runs/{schema}" 
    print(f"Running notebook for schema: {schema}")
    dbutils.notebook.run(
        "/Workspace/Users/depoplimontj@delawareconsulting.com/delaware_lakebridge/.lakebridge/switch/notebooks/00_main",
        timeout_seconds=100000,
        arguments={
            "catalog": catalog,
            "schema": schema,
            "input_dir": input_dir,
            "output_dir": output_folder,
            "source_tech": source_tech,
            "foundation_model": foundation_model,
            "switch_config_path": switch_config_path
        }
    )
    print(f"Completed schema: {schema}")
```
